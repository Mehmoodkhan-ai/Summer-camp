## LOOPS

A loop is something that repeats a set of instructions until a condition is met.

In everyday life

- Brushing each tooth one by one is a loop.
- Walking around a track multiple times is a loop.
- Repeating a song playlist is a loop.

In programming

A loop lets a computer repeat code without writing it over and over.

Write a program that takes a number n as input from the user and generates the first 
n terms of the series formed by squaring the natural numbers. 
- Sample output 
- Enter a number: 6 
- The first 6 terms of the series are: 
- 1 4 9 16 25 36 

```python
num=int(input("HOW MANY YOU WANT TO WISH"))
print(f"The first {num} terms of the series are:")
for temp in range(num):
    print((temp+1)**2)
```
